```python
import random
import numpy as np
import pandas as pd
```

```python
##### Suponiendo que cada cluster tiene observaciones y que estas observaciones pueden estar en uno o más clusters, 
#suponemos que tiene la forma siguiente (simulamos datos para hacer el ejercicio):
observaciones = range(0,200)
clusters = {}

for idx in range(0,10):
    num_obs = random.randrange(5, 40)
    clusters["cluster_"+str(idx)]=random.sample(observaciones,num_obs)
```

```python
clusters
```

```python
##Numero de clusters, estos vienen en formato de diccionario
num_clusters = len(clusters)
```

```python
##Creamos una matriz de adyacencia con puros ceros

matrix_inicial = np.zeros((num_clusters, num_clusters))
names = [cluster for cluster in clusters.keys()]
matrix_distances = pd.DataFrame(matrix_inicial, index=names, columns=names)
matrix_distances
```

```python
##Hacemos una comparacion entre cada cluster, para sacar el numero de elementos que comparten, quiza no es lo más
##eficiente pero es muy sencillo

for idx_1 in clusters.keys():
    for idx_2 in clusters.keys():
        if idx_1 != idx_2:
            #Elementos en cada cluster
            elementos_uno = float(len(clusters[idx_1]))
            elementos_dos = float(len(clusters[idx_2]))
            
            #El cluster con el menor numero de elementos
            minimo_elementos = min(elementos_uno, elementos_dos)
            
            #cuantos elementos comparten cada cluster
            elementos_compartidos = len(list(set(clusters[idx_1]).intersection(clusters[idx_2])))
            
            #Asignamos a la matriz la distancia como 1 - num_elemntos_comp / minim_elementos
            matrix_distances[idx_1][idx_2] = 1- (elementos_compartidos/minimo_elementos)            
```
